```python
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
import pandas as pd
```

```python
consumer=KafkaConsumer("supply_chain",bootstrap_servers=["52.66.235.7:9092"],
                      value_deserializer=lambda x:
                      loads(x.decode("utf-8"))
                      )
```

```python
from s3fs import S3FileSystem
```

```python

s3=S3FileSystem()
```

```python
for count, i in enumerate(consumer):
    print(f"count value: {count}")
    with s3.open(f"s3://supplydata/data/supply_chain_{count}.json",'w') as file :
        json.dump(i.value,file)
    print(f"Stored data for supply ID {count} in S3")    
```
